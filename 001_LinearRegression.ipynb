```python
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Area Population']]
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
ypred = model.predict(x_test)
ypred
```

```python
sns.scatterplot((ypred),y_test)
```

```python
sns.regplot((ypred),y_test)
```

```python
pd.DataFrame(model.coef_,index=x.columns,columns=['coeff'])
```

```python
pd.to_pickle(model,'houseprice.pkl')
```
